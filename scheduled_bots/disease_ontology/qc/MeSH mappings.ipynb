# do - mesh mappings

```python
from rdflib import Graph, URIRef
import pandas as pd

pd.set_option("display.max_rows", None, "display.max_columns", None)

print("\nDownloading the Disease Ontology...")
url = "https://raw.githubusercontent.com/DiseaseOntology/HumanDiseaseOntology/main/src/ontology/releases/2020-11-11/doid.owl"

doGraph = Graph()
doGraph.parse(url, format="xml")

pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(columns=["mesh", "do", "counts"])
qres = doGraph.query("""SELECT ?exactMatch (GROUP_CONCAT(?do;separator="|") as ?mondoUrls) (COUNT(?do) as ?counts)
                        WHERE {
                        
                        
                            ?do skos:exactMatch ?exactMatch .
                            
                            
                             FILTER (REGEX(str(?exactMatch), "mesh", "i"))
                            } GROUP BY ?exactMatch ORDER BY DESC(?counts)""")
for row in qres:
    if int(row[2])>1:
        df = df.append({"mesh": str(row[0]), "do": str(row[1]),"counts":  str(row[2]),}, ignore_index=True)
                     
df 
```

# mesh - mondo mappings

```python
print("\nDownloading Mondo ...")
url = "http://purl.obolibrary.org/obo/mondo.owl"

mondoGraph = Graph()
mondoGraph.parse(url, format="xml")

df = pd.DataFrame(columns=["mesh", "mondo", "counts"])
qres = mondoGraph.query("""PREFIX obo: <http://www.geneontology.org/formats/oboInOwl#>
       PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
       PREFIX oboInOwl: <http://www.geneontology.org/formats/oboInOwl#>
       PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
       SELECT ?exactMatch (GROUP_CONCAT(?mondo;separator="|") as ?mondoUrls) (COUNT(?mondo) as ?counts) WHERE {
           
           
           ?mondo skos:exactMatch ?exactMatch .
           
           
           FILTER (REGEX(str(?exactMatch), "mesh"))
        
       } GROUP BY ?exactMatch ORDER BY DESC(?counts)""")
       
for row in qres:
    if int(row[2])>1:
        df = df.append({"mesh": str(row[0]), "mondo": str(row[1]),"counts":  str(row[2]),}, ignore_index=True)
                     
df
```
