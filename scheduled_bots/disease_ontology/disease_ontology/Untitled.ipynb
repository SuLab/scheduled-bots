```python
from wikidataintegrator import wdi_core, wdi_login
```

```python
query = """
SELECT * WHERE {
  ?item wdt:P486 ?mesh .
  FILTER REGEX(?mesh, ":")
  } """
df = wdi_core.WDFunctionsEngine.execute_sparql_query(query, as_dataframe=True)
```

```python
import pprint

WDUSER = "Andrawaag"
WDPASS = "k1T0BoAUHKTj"

login = wdi_login.WDLogin(WDUSER, WDPASS)

def remove_colon_from_mesh(qid):
    item = wdi_core.WDItemEngine(wd_item_id=qid)
    json = item.get_wd_json_representation()
    if "P486" in json["claims"].keys():
        for claim in json["claims"]["P486"]:
            #pprint.pprint(claim)
            if ":" in claim["mainsnak"]["datavalue"]["value"]:
                print(claim["id"], claim["mainsnak"]["datavalue"]["value"])
                wdi_core.WDItemEngine.delete_statement(statement_id=claim["id"], revision=item.lastrevid, login=login)
```

```python
import time
for index, row in df.iterrows():
    qid = row["item"].replace("http://www.wikidata.org/entity/", "")
    remove_colon_from_mesh(qid)
    time.sleep(10)
    
```
